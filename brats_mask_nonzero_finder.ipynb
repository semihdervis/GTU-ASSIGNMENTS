```python
import nibabel as nib
img = nib.load('/content/drive/MyDrive/BraTS2021_00495/BraTS2021_00495_seg.nii/BraTS2021_00495_seg.nii')
```

```python
import matplotlib.pyplot as plt

data = img.get_fdata()
fig, axes = plt.subplots(nrows=data.shape[2] // 15, ncols=3, figsize=(12, data.shape[2] // 5))

for i, ax in enumerate(axes.flat):
  slice_index = i * 5
  ax.imshow(data[:, :, slice_index], cmap='gray')
  ax.axis('off')
  ax.set_title(f'{slice_index}/{data.shape[2]}')  # Add title with slice index and total slices
plt.show()
```

```python
def has_non_zero_pixel(slice):
  return np.any(slice != 0)
```
